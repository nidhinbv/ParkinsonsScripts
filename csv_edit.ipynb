```python
import pandas as pd
#import numpy as np
import os
```

```python
#Get the name of the file and the directory
inPath=os.path.abspath("C:/Users/nidhi/Downloads/Music/w2096texas weird.csv")
fileName=os.path.basename(inPath)
print(fileName)
```

```python
#Create a dataframe from the CSV
df=pd.read_csv(inPath, encoding="ISO-8859-1")
```

```python
#Count backspaces
backCount=0
for row in df['output']:
    if row=="BACK":
        backCount+=1
print(backCount)

#Make a new column that contains the value 
df['totalBackspaces']=backCount
#df
```

```python
#Delete unnecessary columns
del df['name']
del df['value']
del df['name2']
del df['value3']
df
```

```python
#Delete blank rows
df =df[df.id.notnull()]
#Delete rows that don't involve actual typing 
df =df[df['type']=="keyboard"]
df
```
